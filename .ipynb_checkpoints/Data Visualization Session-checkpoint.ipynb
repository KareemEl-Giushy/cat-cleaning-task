```python
def showplot(col, prop='count',color=None):
    plt.figure(figsize=(30, 10))
    sns.countplot(data=d, x=col, order=d[col].value_counts().index[:15], color=color)
```

```python
showplot('country')
```

```python
year = [
    d[d['arrival_date_year'] == 2015],
    d[d['arrival_date_year'] == 2016],
    d[d['arrival_date_year'] == 2017],
]
fig, ax = plt.subplots(1, 3, figsize=(30,20))
for i, y in enumerate(year):
    fig.sca(ax[i])
    sns.countplot(data=d, y='arrival_date_month')
```
